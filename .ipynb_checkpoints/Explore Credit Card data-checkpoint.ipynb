```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gs
```

```python
df = pd.read_csv("creditcard.csv")
```

Here are the first few observations from the dataset:

```python
df.head()
```

```python
df.describe()
```

Note that there are 31 variables: time, amount, class, and 28 anonymous variables (principle components of credit card transaction features).

First, I look at the time variable:

```python
print("Fraud\n",df.Time[df.Class==1].describe())
print("\nNormal\n",df.Time[df.Class==0].describe())
```

```python
f, (ax1, ax2) = plt.subplots(2,1, sharex=True, figsize=(12,6))

ax1.hist(df.Time[df.Class==1], bins = 50)
ax1.set_title('Fraud')

ax2.hist(df.Time[df.Class ==0], bins = 50)
ax2.set_title('Normal')

plt.xlabel('Time in seconds')
plt.ylabel('Number of transactiosn')
plt.show()
```

```python
print("Fraud\n", df.Time[df.Class==1].describe())
print("\nNormal\n", df.Time[df.Class==0].describe())
```

```python
# Now we draw graphs for the amounts of transactions
f, (ax1, ax2) = plt.subplots(2,1, sharex=True, figsize=(12,4))
ax1.hist(df.Amount[df.Class==1], bins = 20)
ax1.set_title('Fraud')

ax2.hist(df.Amount[df.Class==0], bins=20)
ax2.set_title('Normal')

plt.xlabel('Amount of transactions')
plt.ylabel('Number of transactions')
plt.yscale('log')
plt.show()
```

```python
# here are the histograms of all other 28 anonymous variables
plt.figure(figsize=(12,28*4))
grid = gs.GridSpec(28, 1)
for i, covariate in enumerate(df.ix[:,1:29].columns):
    plt.subplot(grid[i])
    plt.hist(df[covariate], bins = 70)
plt.show()
```
