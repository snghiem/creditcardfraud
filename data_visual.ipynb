```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('creditcard.csv')
df.describe()
```

```python
df['Class'].value_counts()
```

```python
df.head()
```

```python
%matplotlib inline
sns.FacetGrid(df, hue="Class", size=5).map(plt.scatter, "Time", "Amount").add_legend()
```

```python
sns.FacetGrid(df, hue="Class", size=5).map(plt.scatter, "V1", "V2").add_legend()
```

```python
sns.FacetGrid(df, hue="Class", size=5).map(plt.scatter, "V1", "Time").add_legend()
```

```python
sns.FacetGrid(df, hue="Class", size = 6)\
.map(sns.kdeplot, "V1").add_legend()
```

```python
from pandas.tools.plotting import radviz
radviz(df, "Class")
```

```python
# Andrew Curves using attributes of data as coefficients for Fourier Series
from pandas.tools.plotting import andrews_curves
andrews_curves(df, "Class")
```

```python
sns.heatmap(df.corr())
```

```python
import matplotlib.gridspec as gs
# here are the histograms of all other 28 anonymous variables
plt.figure(figsize=(12,28*4))
grid = gs.GridSpec(28, 1)
for i, covariate in enumerate(df.ix[:,1:29].columns):
    plt.subplot(grid[i])
    plt.hist(df[covariate], bins = 70)
plt.show()
```
